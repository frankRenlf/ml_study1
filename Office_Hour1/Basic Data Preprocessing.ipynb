```python
## drop missing data
df1 = df.dropna(subset=["Fare"]) # dropna(how="all/any/...", threshold=2, axis=0, 1)
df1
```

```python
## stats (1.5 IQR, 95% CI)
### 1.5 IQR
q75, q25 = np.percentile(df1["Age"], [75,25])
iqr = q75 - q25
print(iqr, q75, q25)
```

```python
min_citer = q25 - 1.5*iqr
max_citer = q75 + 1.5*iqr
```

```python
df1[df1["Age"] < min_citer]
```

```python
df1[df1["Age"] > max_citer]
```

```python
## 95% C.I
import scipy.stats as st
lower, upper = st.t.interval(0.95, len(df1["Age"])-1, loc=np.mean(df1["Age"]), scale=st.sem(df1["Age"]))
print(lower, upper)
```
